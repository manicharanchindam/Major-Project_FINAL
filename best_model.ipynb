### **Data Preparation**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.impute import SimpleImputer

# Load the dataset
df = pd.read_csv('student_dataset.csv')

# Drop rows with missing values
df.dropna(inplace=True)

# Split the dataset into features (X) and target variable (y)
X = df[['HSGA', 'SAT', 'GAT']]
y = df['CGPA_Category']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize a dictionary to store evaluation metrics
eval_metrics = {}
```

```python
from sklearn.neural_network import MLPClassifier

# Initialize the MLPClassifier (Artificial Neural Network) model
ann_model = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=500, random_state=42)

# Train the model
ann_model.fit(X_train, y_train)

# Predict the CGPA categories
ann_y_pred = ann_model.predict(X_test)

# Calculate metrics
eval_metrics["Artificial Neural Network"] = {
    "Model": "Artificial Neural Network",
    "Accuracy": accuracy_score(y_test, ann_y_pred),
    "Precision": precision_score(y_test, ann_y_pred, average='weighted'),
    "Recall": recall_score(y_test, ann_y_pred, average='weighted'),
    "F1 Score": f1_score(y_test, ann_y_pred, average='weighted')
}
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the model
dt_model = DecisionTreeClassifier(random_state=42)

# Train the model
dt_model.fit(X_train, y_train)

# Predict the CGPA categories
dt_y_pred = dt_model.predict(X_test)

# Calculate metrics
eval_metrics["Decision Tree (J48)"] = {
    "Model": "Decision Tree",
    "Accuracy": accuracy_score(y_test, dt_y_pred),
    "Precision": precision_score(y_test, dt_y_pred, average='weighted'),
    "Recall": recall_score(y_test, dt_y_pred, average='weighted'),
    "F1 Score": f1_score(y_test, dt_y_pred, average='weighted')
}

```

```python
from sklearn.svm import SVC

# Initialize the SVM model
svm_model = SVC(kernel='rbf', random_state=42)

# Train the model
svm_model.fit(X_train, y_train)

# Predict the CGPA categories
svm_y_pred = svm_model.predict(X_test)

# Calculate metrics
eval_metrics["Support Vector Machine"] = {
    "Model": "Support Vector Machine",
    "Accuracy": accuracy_score(y_test, svm_y_pred),
    "Precision": precision_score(y_test, svm_y_pred, average='weighted'),
    "Recall": recall_score(y_test, svm_y_pred, average='weighted'),
    "F1 Score": f1_score(y_test, svm_y_pred, average='weighted')
}
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the model
nb_model = GaussianNB()

# Train the model
nb_model.fit(X_train, y_train)

# Predict the CGPA categories
nb_y_pred = nb_model.predict(X_test)

# Calculate metrics
eval_metrics["Gaussian Naive Bayes"] = {
    "Model": "Gaussian Naive Bayes",
    "Accuracy": accuracy_score(y_test, nb_y_pred),
    "Precision": precision_score(y_test, nb_y_pred, average='weighted'),
    "Recall": recall_score(y_test, nb_y_pred, average='weighted'),
    "F1 Score": f1_score(y_test, nb_y_pred, average='weighted')
}
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the model
gb_model = GradientBoostingClassifier(random_state=42)

# Train the model
gb_model.fit(X_train, y_train)

# Predict the CGPA categories
gb_y_pred = gb_model.predict(X_test)

# Calculate metrics
eval_metrics["Gradient Boosting"] = {
    "Model": "Gradient Boosting",
    "Accuracy": accuracy_score(y_test, gb_y_pred),
    "Precision": precision_score(y_test, gb_y_pred, average='weighted'),
    "Recall": recall_score(y_test, gb_y_pred, average='weighted'),
    "F1 Score": f1_score(y_test, gb_y_pred, average='weighted')
}
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the model
knn_model = KNeighborsClassifier()

# Train the model
knn_model.fit(X_train, y_train)

# Predict the CGPA categories
knn_y_pred = knn_model.predict(X_test)

# Calculate metrics
eval_metrics["K Nearest Neighbors"] = {
    "Model": "K Nearest Neighbors",
    "Accuracy": accuracy_score(y_test, knn_y_pred),
    "Precision": precision_score(y_test, knn_y_pred, average='weighted'),
    "Recall": recall_score(y_test, knn_y_pred, average='weighted'),
    "F1 Score": f1_score(y_test, knn_y_pred, average='weighted')
}
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the model
rf_model = RandomForestClassifier(random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Predict the CGPA categories
rf_y_pred = rf_model.predict(X_test)

# Calculate metrics
eval_metrics["Random Forest"] = {
    "Model": "Random Forest",
    "Accuracy": accuracy_score(y_test, rf_y_pred),
    "Precision": precision_score(y_test, rf_y_pred, average='weighted'),
    "Recall": recall_score(y_test, rf_y_pred, average='weighted'),
    "F1 Score": f1_score(y_test, rf_y_pred, average='weighted')
}
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the model
lr_model = LogisticRegression(max_iter=1000, random_state=42)

# Train the model
lr_model.fit(X_train, y_train)

# Predict the CGPA categories
lr_y_pred = lr_model.predict(X_test)

# Calculate metrics
eval_metrics["Logistic Regression"] = {
    "Model": "Logistic Regression",
    "Accuracy": accuracy_score(y_test, lr_y_pred),
    "Precision": precision_score(y_test, lr_y_pred, average='weighted'),
    "Recall": recall_score(y_test, lr_y_pred, average='weighted'),
    "F1 Score": f1_score(y_test, lr_y_pred, average='weighted')
}
```

```python
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
from colorama import Fore, Style

# Set figure size
plt.figure(figsize=(8, 20))

# Bar graph for Precision comparison
plt.subplot(4, 1, 1)
precisions = [metrics['Precision'] for metrics in eval_metrics.values()]
plt.bar(eval_metrics.keys(), precisions, color=plt.cm.Paired(np.arange(len(eval_metrics))))
plt.xlabel('Model')
plt.ylabel('Precision')
plt.title('Precision Comparison')
plt.xticks(rotation=45)

# Line graph for Recall comparison
plt.subplot(4, 1, 2)
recalls = [metrics['Recall'] for metrics in eval_metrics.values()]
plt.plot(eval_metrics.keys(), recalls, marker='o', color='b')
plt.xlabel('Model')
plt.ylabel('Recall')
plt.title('Recall Comparison')
plt.xticks(rotation=45)

# Stacked bar graph for F1 Score comparison
plt.subplot(4, 1, 3)
f1_scores = [metrics['F1 Score'] for metrics in eval_metrics.values()]
plt.bar(eval_metrics.keys(), f1_scores, color=plt.cm.Paired(np.arange(len(eval_metrics))))
plt.xlabel('Model')
plt.ylabel('F1 Score')
plt.title('F1 Score Comparison')
plt.xticks(rotation=45)

# Pie chart for Accuracy comparison
plt.subplot(4, 1, 4)
sizes = [metrics['Accuracy'] for metrics in eval_metrics.values()]
plt.pie(sizes, labels=eval_metrics.keys(), autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired(np.arange(len(eval_metrics))))
plt.title('Accuracy Comparison')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()

# Display metrics in a colorful table
headers = [Fore.BLUE + 'Model', 'Precision', 'Recall', 'F1 Score', 'Accuracy' + Style.RESET_ALL]
table_data = []
for model, metrics in eval_metrics.items():
    row = [Fore.GREEN + model, str(metrics['Precision']), str(metrics['Recall']), str(metrics['F1 Score']), str(metrics['Accuracy']) + Style.RESET_ALL]
    table_data.append(row)

print(tabulate(table_data, headers=headers, tablefmt='grid'))
```
